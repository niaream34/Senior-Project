```python
import os, glob
from pathlib import Path

IMG_ROOT = Path(extract_path) / "Images"
SPLIT    = "trainval"       # or "test-dev"
OPT_DIR  = IMG_ROOT / SPLIT / "Optical"
TH_DIR   = IMG_ROOT / SPLIT / "Thermal"

def list_sequences(modal_dir):
    return sorted([p.name for p in modal_dir.iterdir() if p.is_dir()])

def list_pairs(seq_name):
    opt_seq = OPT_DIR / seq_name
    th_seq  = TH_DIR  / seq_name
    opt_frames = sorted(glob.glob(str(opt_seq/"*.jpg")))
    th_frames  = sorted(glob.glob(str(th_seq/"*.jpg")))
    pairs = []
    for of in opt_frames:
        idx = Path(of).stem
        tf = th_seq / f"{idx}.jpg"
        if tf.exists():
            pairs.append((of, str(tf)))
    return pairs

all_seqs = sorted(set(list_sequences(OPT_DIR)).intersection(set(list_sequences(TH_DIR))))
print(f"ðŸ”Ž found {len(all_seqs)} sequences (opt & thermal matched)")
print(all_seqs[:10])
```

```python
import numpy as np, cv2
from skimage.morphology import remove_small_objects
from skimage.measure import label, regionprops

def ensure_dir(p): Path(p).mkdir(parents=True, exist_ok=True)

def overlay_mask(img_bgr, mask, alpha=0.45, color=(0,0,255)):
    color_img = np.zeros_like(img_bgr); color_img[mask>0] = color
    return cv2.addWeighted(img_bgr, 1.0, color_img, alpha, 0)

def clean_keep_largest(mask, min_area=1000):
    lbl = label(mask>0, connectivity=2)
    if lbl.max()==0: return np.zeros_like(mask, np.uint8)
    areas = [(r.area, r.label) for r in regionprops(lbl)]
    areas.sort(reverse=True)
    out = (lbl==areas[0][1]).astype(np.uint8)
    out = cv2.morphologyEx(out, cv2.MORPH_CLOSE, np.ones((7,7), np.uint8))
    return out

# simple, robust registration (feature-based; can be skipped in fast mode)
def register_thermal_to_opt(opt_bgr, th_bgr):
    o = cv2.cvtColor(opt_bgr, cv2.COLOR_BGR2GRAY)
    t = cv2.cvtColor(th_bgr,  cv2.COLOR_BGR2GRAY)
    h, w = o.shape[:2]
    orb = cv2.ORB_create(1500)
    ko, do = orb.detectAndCompute(o, None)
    kt, dt = orb.detectAndCompute(t, None)
    if do is None or dt is None or len(ko)<10 or len(kt)<10:
        return cv2.resize(th_bgr, (w, h))
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(dt, do)
    matches = sorted(matches, key=lambda x: x.distance)[:200]
    if len(matches) < 8: return cv2.resize(th_bgr, (w, h))
    pts_t = np.float32([kt[m.queryIdx].pt for m in matches]).reshape(-1,1,2)
    pts_o = np.float32([ko[m.trainIdx].pt for m in matches]).reshape(-1,1,2)
    H, _ = cv2.estimateAffinePartial2D(pts_t, pts_o, method=cv2.RANSAC, ransacReprojThreshold=3.0)
    if H is None: return cv2.resize(th_bgr, (w, h))
    return cv2.warpAffine(th_bgr, H, (w, h))
```

```python
def blade_mask_from_thermal(th_bgr):
    th = cv2.cvtColor(th_bgr, cv2.COLOR_BGR2GRAY)
    th = cv2.GaussianBlur(th, (5,5), 0)
    bg = cv2.morphologyEx(th, cv2.MORPH_OPEN, np.ones((31,31), np.uint8))
    th_bg = cv2.subtract(th, bg)
    _, m = cv2.threshold(th_bg, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return clean_keep_largest(m, 1000)

def blade_mask_from_optical(opt_bgr):
    hsv = cv2.cvtColor(opt_bgr, cv2.COLOR_BGR2HSV)
    v = cv2.GaussianBlur(hsv[:,:,2], (5,5), 0)
    m = cv2.adaptiveThreshold(v,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 51, 2)
    edges = cv2.Canny(v, 60, 120)
    edges = cv2.dilate(edges, np.ones((3,3), np.uint8), 1)
    m = cv2.bitwise_or(m, edges)
    m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, np.ones((9,9), np.uint8))
    return clean_keep_largest(m, 1000)

def fuse_blade_masks(opt_bgr, th_bgr):
    th_mask = blade_mask_from_thermal(th_bgr)
    opt_mask = blade_mask_from_optical(opt_bgr)
    fused = cv2.bitwise_or(th_mask, opt_mask)
    fused = clean_keep_largest(fused, 2000)
    return fused, opt_mask, th_mask
```

```python
from sklearn.mixture import GaussianMixture

def damage_threshold(th_bgr, blade_mask, manual_thresh=None):
    th = cv2.cvtColor(th_bgr, cv2.COLOR_BGR2GRAY)
    th = cv2.GaussianBlur(th, (5,5), 0)
    bg = cv2.morphologyEx(th, cv2.MORPH_OPEN, np.ones((31,31), np.uint8))
    th_bg = cv2.subtract(th, bg)
    roi = th_bg.copy()
    roi[blade_mask==0] = 0
    if manual_thresh is None:
        _, dmg = cv2.threshold(roi, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    else:
        _, dmg = cv2.threshold(roi, manual_thresh, 255, cv2.THRESH_BINARY)
    dmg = cv2.morphologyEx(dmg, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))
    dmg = remove_small_objects(dmg.astype(bool), min_size=50).astype(np.uint8)*255
    return dmg

def damage_gmm(th_bgr, blade_mask, max_k=5, sample_pixels=None):
    th = cv2.cvtColor(th_bgr, cv2.COLOR_BGR2GRAY)
    th = cv2.GaussianBlur(th, (5,5), 0)
    bg = cv2.morphologyEx(th, cv2.MORPH_OPEN, np.ones((31,31), np.uint8))
    th_bg = cv2.subtract(th, bg)
    idx = np.where(blade_mask>0)
    vals_full = th_bg[idx].reshape(-1,1).astype(np.float32)
    n = vals_full.shape[0]
    if n == 0: return np.zeros_like(th_bg, np.uint8)
    vals = vals_full
    if sample_pixels and n > sample_pixels:
        sel = np.random.choice(n, sample_pixels, replace=False)
        vals = vals_full[sel]
    best_g, best_bic = None, 1e18
    for k in range(2, max_k+1):
        g = GaussianMixture(n_components=k, covariance_type='full', random_state=0)
        g.fit(vals)
        bic = g.bic(vals)
        if bic < best_bic: best_bic, best_g = bic, g
    lab = best_g.predict(vals_full)
    means = best_g.means_.ravel()
    dmg_comp = np.argmax(means)         # hottest = damage
    dmg = np.zeros_like(th_bg, np.uint8)
    dmg[idx[0], idx[1]] = (lab==dmg_comp).astype(np.uint8)*255
    dmg = cv2.morphologyEx(dmg, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))
    dmg = remove_small_objects(dmg.astype(bool), min_size=50).astype(np.uint8)*255
    return dmg
```

```python
def compute_enthalpy(th_bgr, dmg_mask, rho, cp, ts_m,
                     pixel_size_m=None, use_region_lw=False, region_lw_m=(0.10, 0.02),
                     Ta=0.0):
    """
    Î”h_j = Ï * cp * V_pix * sum_i (T_i,j - T_a)
    T_i,j = min(T_j) + G_ij * Î”T_j ;  Î”T_j = (max(T_j)-min(T_j))/255
    """
    gray = cv2.cvtColor(th_bgr, cv2.COLOR_BGR2GRAY).astype(np.float32)
    Tmin, Tmax = gray.min(), gray.max()
    dT = (Tmax - Tmin) / 255.0 if Tmax > Tmin else 0.0
    T = Tmin + gray * dT
    dmg = (dmg_mask > 0)
    if not dmg.any(): return 0.0

    if not use_region_lw:
        assert pixel_size_m is not None, "Set pixel_size_m (m/pixel) or use use_region_lw=True."
        Vpix = ts_m * (pixel_size_m**2)
        return float(rho * cp * Vpix * np.sum(T[dmg] - Ta))

    # per-component Vpix using Vpix = ts*l*w/(Nd_l * Nd_w)
    from skimage.measure import label, regionprops
    lbl = label(dmg.astype(np.uint8), connectivity=2)
    l_m, w_m = region_lw_m
    total = 0.0
    for r in regionprops(lbl):
        Nd_l = max(1, r.bbox[2] - r.bbox[0])
        Nd_w = max(1, r.bbox[3] - r.bbox[1])
        Vpix = ts_m * l_m * w_m / (Nd_l * Nd_w)
        coords = r.coords
        total += rho * cp * Vpix * np.sum(T[coords[:,0], coords[:,1]] - Ta)
    return float(total)
```

```python
def false_color_thermal(th_bgr):
    g = cv2.cvtColor(th_bgr, cv2.COLOR_BGR2GRAY)
    return cv2.applyColorMap(g, cv2.COLORMAP_JET)

def draw_damage_contours(base_bgr, dmg_mask, color=(0,255,255), thickness=2):
    cnts, _ = cv2.findContours(dmg_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    out = base_bgr.copy()
    cv2.drawContours(out, cnts, -1, color, thickness)
    return out

def make_panel(opt_bgr, th_bgr, blade_mask, dmg_mask, title=None):
    opt_overlay = overlay_mask(opt_bgr, blade_mask, alpha=0.35)
    th_color    = overlay_mask(false_color_thermal(th_bgr), blade_mask, alpha=0.35)
    opt_cont = draw_damage_contours(opt_overlay, dmg_mask)
    th_cont  = draw_damage_contours(th_color,  dmg_mask)
    fused = cv2.addWeighted(opt_bgr, 0.6, false_color_thermal(th_bgr), 0.4, 0)
    fused = overlay_mask(fused, dmg_mask, 0.45, color=(0,0,255))
    panel = np.hstack([opt_cont, th_cont, fused])
    if title:
        cv2.putText(panel, title, (30,40), cv2.FONT_HERSHEY_SIMPLEX, 1.1, (255,255,255), 2, cv2.LINE_AA)
    return panel
```

```python
import csv, math
from tqdm import tqdm

# --- FAST MODE toggles ---
FAST_DOWNSCALE   = 0.6         # 1.0 = full-res; 0.5â€“0.7 is usually fine
FAST_DISABLE_REG = True        # skip feature registration (big speed win)
FAST_SAVE_VIS    = True        # turn off while debugging
FAST_MAX_FRAMES  = None        # e.g., 10 to test quickly
GMM_SAMPLE_PX    = 10000       # subsample for GMM fitting (still labels all pixels)

def _downscale(opt, th, scale):
    if scale==1.0: return opt, th
    nh, nw = int(opt.shape[0]*scale), int(opt.shape[1]*scale)
    return cv2.resize(opt,(nw,nh),cv2.INTER_AREA), cv2.resize(th,(nw,nh),cv2.INTER_AREA)

def _reg(opt, th):
    if FAST_DISABLE_REG:
        h,w = opt.shape[:2]; return cv2.resize(th,(w,h))
    return register_thermal_to_opt(opt, th)

def process_sequence(seq_name,
                     out_dir="./outputs_fast",
                     method="gmm", manual_thresh=None,
                     rho=1600.0, cp=900.0, ts_m=0.010,
                     pixel_size_m=0.001, Ta=0.0):
    ensure_dir(out_dir)
    pairs = list_pairs(seq_name)
    if FAST_MAX_FRAMES: pairs = pairs[:FAST_MAX_FRAMES]
    csv_path = Path(out_dir)/f"{seq_name}_{method}_enthalpy.csv"
    with open(csv_path, "w", newline="") as f:
        w = csv.writer(f); w.writerow(["frame","enthalpy_J"])
        for of, tf in tqdm(pairs, desc=f"{seq_name} ({method})"):
            opt = cv2.imread(of); th = cv2.imread(tf)
            opt, th = _downscale(opt, th, FAST_DOWNSCALE)
            th_reg  = _reg(opt, th)
            blade_mask, _, _ = fuse_blade_masks(opt, th_reg)
            if method=="gmm":
                dmg = damage_gmm(th_reg, blade_mask, sample_pixels=GMM_SAMPLE_PX)
            else:
                dmg = damage_threshold(th_reg, blade_mask, manual_thresh=manual_thresh)
            dH = compute_enthalpy(th_reg, dmg, rho, cp, ts_m, pixel_size_m=pixel_size_m, Ta=Ta)
            w.writerow([Path(of).stem, f"{dH:.3f}"])
            if FAST_SAVE_VIS:
                panel = make_panel(opt, th_reg, blade_mask*255, dmg, title=f"{seq_name} | {Path(of).stem}")
                cv2.imwrite(str(Path(out_dir)/f"{seq_name}_{Path(of).stem}_{method}.jpg"), panel)
    print("âœ… Saved CSV:", csv_path)
    if FAST_SAVE_VIS: print("ðŸ–¼ï¸ Saved images to:", out_dir)
```

```python
SEQ = all_seqs[0]     # pick any from the printed list (e.g., "DJI_0101")
print("Testing sequence:", SEQ)

# automated (GMM+BIC)
process_sequence(SEQ, method="gmm", out_dir="./outputs_fast",
                 rho=1600.0, cp=900.0, ts_m=0.010, pixel_size_m=0.001, Ta=0.0)

# expert-tunable threshold (Otsu auto by default)
process_sequence(SEQ, method="thresh", out_dir="./outputs_fast",
                 rho=1600.0, cp=900.0, ts_m=0.010, pixel_size_m=0.001, Ta=0.0)
```
